```python
import pandas as pd

# List of dataset files
files = [
    "Cheapestelectriccars-EVDatabase.csv",
    "electric_vehicle_charging_station_list.csv",
    "ElectricCarData_Clean.csv",
    "EVIndia.csv"
]

for file in files:
    print("\nProcessing:", file)

    # Load dataset
    df = pd.read_csv(file)

    # Show first rows
    print("\nFirst 5 rows:")
    print(df.head())

    # Dataset information
    print("\nDataset Info:")
    print(df.info())

    # Check missing values
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Remove missing values
    df = df.dropna()

    # Check duplicates
    print("\nDuplicate Rows:", df.duplicated().sum())

    # Remove duplicates
    df = df.drop_duplicates()

    # Clean column names
    df.columns = df.columns.str.lower().str.replace(" ", "_")

    # Dataset statistics
    print("\nDataset Statistics:")
    print(df.describe())

    # Save cleaned dataset
    clean_name = "clean_" + file
    df.to_csv(clean_name, index=False)

    print("\nSaved cleaned dataset as:", clean_name)

print("\nAll datasets cleaned and ready for Tableau visualization.")
```
